```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['newspaper'] + df['radio']
df
```

```python
#creating scatterplot with seaborn
sns.scatterplot(data=df,x=df['total_spend'],y=df['sales'])
```

```python
X = df['total_spend']
y = df['sales']
```

```python
#y = mx + b
#y = B1x + B0
beta_values = np.polyfit(X,y,deg=1)
```

```python
potential_spend = np.linspace(0,500,100)
predicted_sales = beta_values[0]*potential_spend + beta_values[1]
```

```python
sns.scatterplot(x=df['total_spend'],y=df['sales'])
plt.plot(potential_spend,predicted_sales)
```

```python
expected_spend = 200
predicted_sales = beta_values[0] * expected_spend + beta_values[1]
predicted_sales
```

```python
beta_values = np.polyfit(X,y,deg=3)
beta_values
```

```python
#y = B3x**3 + B2x**2 + B1x + B0
pot_spend = np.linspace(0,500,100)
predicted_sales = (beta_values[0]*(pot_spend**3)) + (beta_values[1]*(pot_spend**2)) + (beta_values[2]*(pot_spend)) + beta_values[3]
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(pot_spend,predicted_sales,color='red')
```
