<h2>Figure Object</h2>
<ul>
<li>We create a figure object , then add axes to those figure object</li>
<li>Then we plot on those axis</li>
<li>This allows for very robust controls over the plot</li>
    <li><b>Note:</b><br>     
    &nbsp The figure object is not visible unless you add axes to it</li>
</ul>

```python
#create a figure object default size 720x720
fig = plt.figure()
#adding axes
axes = fig.add_axes([0,0,1,1]) #adding axis first two values of array correspond to the lower left corner relative to the figure
#second set of parameters are the width and height , 1 means 100% 2 means 100%
axes.plot(x,y)#plotting the values on the axis
```

```python
#data for plotting
a = np.linspace(0,10,11)
b = a ** 4
```

```python
b
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
x
```

```python
y
```

```python
#create a blank canvas
fig = plt.figure()
#add axes
axes = fig.add_axes([0,0,1,1])
#create a plot of values on the axes
axes.plot(x,y);
```

```python
fig = plt.figure()
#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_ylabel("B")
axes1.set_xlabel("A")
axes1.set_title("Power of 4")
#small axes
axes2 = fig.add_axes([1.5,1,0.2,0.2])
axes2.set_xlim(1,2)
axes2.set_ylim(0,100)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")
axes2.plot(a,b)
#
```

```python
#Adding dpi
fig = plt.figure(dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

### Subplots

```python
a = np.linspace(0,10,11)
b = a ** 4
x = np.arange(2,10)
y = x * 2
```

```python
fig , axes = plt.subplots(2,2)
axes[1][0].plot(x,y)
```

```python
fig , axes = plt.subplots(2,2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)
fig.subplots_adjust(wspace=0.3,hspace=0.2)
fig.suptitle("Figure Level Title")
fig.dpi = 800
fig.savefig("subplot.png",dpi=300)
```
